```python
outline_generator = autogen.AssistantAgent(
    name="Outline_Generator", llm_config=llm_config,
    system_message=f"""Outline Generator. Creates an outline for an interesting newspaper article for {newspaper_name} newspaper given the topic by the user.
    There should be 1 intriguing title and bulletpoints of topics to cover in order. The title should be attention-triggeting and engaging."""
)
writer = autogen.AssistantAgent(
    name="Writer", llm_config=llm_config,
    system_message=f"""Writer. Writes an interesting newspaper article for {newspaper_name} online newspaper using the outline created by the Outline Generator for the user's topics.
    Keep the title and subtitle structure that Outline Generator created, but change bullet points to an engaging paragraph. Format should look like:
    **Title**

    *Subtitle*
    content
    ...
    After every feedback by the Editor, Stylist and Newspaper Style Checker, you write the article again based on their feedback."""
)

stylist = autogen.AssistantAgent(
    name="Stylist", llm_config=llm_config,
    system_message="""Stylist. Gives short feedback about the structure of the Writer's article to maintain engagement and curiosity, using questions, action words, and numbers in subheadings.
    Be very brief and concise with your feedback. Max 100 words. Be very specific and point to a specific part of the article for changes.
    If the writer's article looks satisfactory  and no changes to be made, just write "looks good"."""
)
editor = autogen.AssistantAgent(
    name="Editor", llm_config=llm_config,
    system_message="""Editor. Gives short feedback about the Writer's article to make sure it has a nice flow and a suitable tone for a newspaper. It should be professional.
    Be very brief and concise with your feedback. Max 100 words. Be very specific and point to a specific part of the article for changes.
    If the writer's article looks satisfactory  and no changes to be made, just write "looks good"."""
)


user_proxy = autogen.UserProxyAgent(
    name="User_Proxy",
    human_input_mode="TERMINATE",
    # is_termination_msg=lambda x: x.get("content", "") and x.get("content", "").rstrip().endswith("TERMINATE"),
    llm_config=llm_config,
    # max_consecutive_auto_reply=3,
    is_termination_msg=lambda x: x.get("content", "").rstrip().endswith("TERMINATE"),
    # code_execution_config=False,
    code_execution_config={"use_docker": False},
    system_message = "User Proxy. When Editor, Stylist and Newspaper Style Checker have positive feedbacks and the article generation process is done, print TERMINATE. If there are improvements to be made, then just print CONTINUE."
#     system_message="""User Proxy. Provides feedback on the article's effectiveness in engaging readers in the local newspaper and 
# Ensures the article meets overall objectives and resonates with the intended audience.
# Relays audience and management preferences to the team for necessary adjustments.
# Once you have positive feedback from Writer, Narrative designer and Style specialist, print TERMINATE to show task is done."""
)

groupchat = autogen.GroupChat(agents=[user_proxy, outline_generator, writer, newspaper_style_check, stylist, editor], messages=[], max_round=15)
manager = autogen.GroupChatManager(groupchat=groupchat, llm_config=llm_config,
                                   system_message="First call the Outline Generator, then the writer. Then call, in this order, Stylist, Editor, Newspaper Style Checker, User Proxy and Writer, repeatedly. Call the same agent ONLY once in a row.")
```

```python
TOPICS = {"History of Microsoft and OpenAI partnership",
          "OpenAI release Sora text-to-video model that generates high quality videos within seconds with natural language",
          "Connect the partnership history to newest innovation from them such as Sora"}

message = f"""The topic is about: """ + str(TOPICS)

user_proxy.initiate_chat(
    manager, message=message
)
```

Add function calling and provide specific "tools" to agents in order to perform more complex tasks. In this case, the task is to extract current trending topics based on country and year.

```python
from typing import Literal
from typing_extensions import Annotated
from pydantic import BaseModel, Field
from pytrends.request import TrendReq

@user_proxy.register_for_execution()
@outline_generator.register_for_llm(description="Trend topics identifier by country and year.")
def trend_by_country_year(
    country: Annotated[str, "Country for which the topic trends will be identified. Can be united states, germany, etc."] = "united states",
    country_code: Annotated[str, "Country code for which the topic trends will be identified. Can be DE, US, GLOBAL (if no country defined) etc."] = "GLOBAL",
    year: Annotated[str, "Year for which the topic trends will be identified"] = "2023",
    realtime: Annotated[int, "Value 1 if user wants real-time / current trends."] = 0,
    yearly: Annotated[int, "Value 1 if user wants general trends by specific year."] = 0,
    # base_currency: Annotated[CurrencySymbol, "Base currency"] = "USD",
    # quote_currency: Annotated[CurrencySymbol, "Quote currency"] = "EUR",
) -> str:
    # quote_amount = exchange_rate(base_currency, quote_currency) * base_amount
    pytrends = TrendReq(hl='en-US', tz=360)
    if realtime==1:
        trends = pytrends.trending_searches(pn=country.lower())
    elif yearly==1:
        trends = pytrends.top_charts(year, hl='en-US', tz=300, geo=country_code)
    return str(trends)
```
